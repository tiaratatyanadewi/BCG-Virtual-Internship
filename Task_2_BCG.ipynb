## Data Visualization

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:

        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')
```

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(['churn']).count()
churn_percentage = churn_total/churn_total.sum()*100
plot_stacked_bars(churn_percentage.transpose(), "Churning status", (5, 5), legend_='lower right')
```

```python
consumption = client_df[['id', 'cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'has_gas', 'churn']]

fig, axs = plt.subplots(nrows=1, figsize=(18, 5))

plot_distribution(consumption, 'cons_12m', axs)
plt.grid()
```

```python
fig, axs = plt.subplots(nrows=1, figsize=(18, 5))
plot_distribution(consumption, 'churn', axs)
plt.grid()
```

```python
churn_channel_sales = client_df.groupby(['channel_sales', 'churn']).size().unstack()
churn_channel_sales_percentage = churn_channel_sales.div(churn_channel_sales.sum(axis=1), axis=0) * 100

plot_stacked_bars(churn_channel_sales_percentage, "Churning status by channel_sales", (20,8), legend_="lower right")

churn_channel_sales_percentage
```

```python
#@title Merged Dataset
df_merged = pd.merge(price_df, client_df, on='id')
df_merged.head()
```

```python
df_merged.info()
```

```python
for col_name in df_merged.select_dtypes("object"):
    df_merged[col_name], _ = df_merged[col_name].factorize()
```

```python
df_merged.info()
```

```python
corr = df_merged.corr()
corr
```

```python
#@title Plot correlation matrix using Heatmap
plt.figure(figsize=(20, 20))
sns.heatmap(corr, annot=True, fmt='.2f')
```

```python
#@title Plot high correlation features
plt.figure(figsize=(20, 20))
sns.heatmap(corr[(corr >= 0.75) | (corr <= -0.65)],
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True)
```

```python
# Define the threshold
positive_threshold = 0.75
negative_threshold = -0.65

# Iterate through the correlation matrix and print only significant correlations
for i in range(len(corr.columns)):
    for j in range(i):
        correlation = corr.iloc[i, j]
        if (correlation > positive_threshold or correlation <= negative_threshold) and not pd.isna(correlation):
            print(f"Correlation between {corr.columns[i]} and {corr.columns[j]}: {correlation:.2f}")
```
